```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import graphviz
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn import metrics
from sklearn import datasets
```

```python
df=pd.read_csv("files folder\\Titanic.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.drop(['cabin','body'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df.survived.fillna((df["survived"].mode()[0]),inplace=True)
df.sex.fillna((df["sex"].mode()[0]),inplace=True)
df.sibsp.fillna((df["sibsp"].median()),inplace=True)
df.pclass.fillna((df["pclass"].mode()[0]),inplace=True)
df.parch.fillna((df["parch"].mode()[0]),inplace=True)
df.ticket.fillna(df["ticket"].mode()[0],inplace=True)
df.fare.fillna((df["fare"].mean()),inplace=True)
df["home.dest"].fillna((df["home.dest"].mode()[0]),inplace=True)
df["age"].fillna(df["age"].mean(),inplace=True)
df["embarked"].fillna(df.embarked.mode()[0],inplace=True)
df.boat.fillna("None",inplace=True)
df.isna().sum()
```

```python
df.dropna(axis=0,how='any',inplace=True)
df.isnull().sum()
```

```python
df.embarked = df.embarked.map({'Q':0, 'C':1,'S':2})
df["sex"]=encoder.fit_transform(df["sex"])
df
```

```python
y=df["survived"] 
x= df[[ "sex", "age"]]
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
classifier=tree.DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
predicted=classifier.predict(x_test)
print("score : {}".format(classifier.score(x_train,y_train)))
```
