```python
a=pd.get_dummies(data["sex"])
data=data.drop("sex",axis=1)
```

```python
df=pd.concat([data,a], axis=1)
df
```

```python
df=df.drop("female", axis = 1)
df.rename(columns={'male':'sex'}, inplace=True)
```

```python
df=df.drop(1309, axis=0)
df
```

```python
df.isnull().sum()
```

```python
df['embarked']=df["embarked"].fillna('S')
df
```

```python
b=pd.get_dummies(df["embarked"])
b
```

```python
df=df.drop("embarked", axis = 1)
df=pd.concat([df,b], axis=1)
df
```

```python
new = df["name"].str.split(",", n = 1, expand = True) 
m=new[1].str.split(".", n = 1, expand = True)
df['title']= m[0]
df.drop(columns =["name"], inplace = True) 
df
```

```python
df['age']=df["age"].fillna(df["age"].mean())
df
```

the title section is easier to analyse

```python
def plot_correlation_map( df ):
    corr = df.corr() #find the corrolation among the columns
    s , ax = plt.subplots( figsize =( 12 , 10 ) ) #created the figure where the values of the corr() will be sorted
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True ) # creates a palette of colors
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 },  #regulates the legend bar of the colors
        ax=ax, 
        annot = True, # writes the values in each box
        annot_kws = { 'fontsize' : 12 } #regulates the police and size of the legend in teh boxes
        )
plot_correlation_map(df)
```
